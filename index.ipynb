### Let's do Regression 

Right here is the real deal. Let's perform a number of simple regression experiments between the chosen independent variables and the dependent variable (price). We shall do this is a loop and in every iteration, we shall pick one of the independent variables  perform following steps:

* Run a simple OLS regression between independent and dependent variables
* Plot a regression line on the scatter plots
* Plot the residuals using `sm.graphics.plot_regress_exog()`.
* Plot a Q-Q plot for regression residuals normality test 
* Store following values in array for each iteration:
    * Independent Variable
    * r_squared'
    * intercept'
    * 'slope'
    * 'p-value'
    * 'normality (JB)' 
* Comment on each output 

```python
# Your code here
import numpy as np
from statsmodels.formula.api import ols
import statsmodels.stats.api as sms
stored_values=[['Independent Variable','r-squared','y-intercept','slope','p-value','JB normality']]
for i in ['crim','dis','rm','zn','age']:
    f='medv~'+i
    model=ols(formula=f,data=df).fit()
    i_min=min(df[i])
    i_max=max(df[i])
    x_new={i:[i_min,i_max]}
    x_new2=[i_min,i_max]
    pred=model.predict(x_new)
    plt.scatter(df[i],df.medv, label=i)
    plt.plot(x_new2,pred,color='red')
    plt.title(i)
    fig = plt.figure(figsize=(15,8))
    fig = sm.graphics.plot_regress_exog(model, i, fig=fig)
    fig = sm.graphics.qqplot(model.resid,line='45',fit=True, dist=stats.norm)
    plt.show()
    r_sq=model.rsquared
    b=model.params[0]
    m=model.params[1]
    p=model.pvalues[1]
    jb=sms.jarque_bera(model.resid)[0]
    stored_values.append([i,r_sq,b,m,p,jb])
```

```python
pd.DataFrame(stored_values)
```

```python
#Your obervations here 
# none of the variables are very predictive of the median value
# rm is the best with an r squared of .48 but even that is not very good
```

So clearly the results are not highly reliable. the best good of fit i.e. r-squared is witnessed with `rm`. So clearly in this analysis this is our best predictor.
